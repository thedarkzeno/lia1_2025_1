## 2. Carregar e Exibir os Dados da Fonte

```python
# -*- coding: utf-8 -*-

# importar a biblioteca usada para trabalhar com vetores e matrizes
import numpy as np

# importar a biblioteca usada para trabalhar com dataframes (dados em formato de tabela) e análise de dados
import pandas as pd 

# importar as bibliotecas usadas para geração de gráficos
import seaborn as sns
import matplotlib.pyplot as plt

print('Bibliotecas carregadas com sucesso!')
```

```python
df_2020_01= pd.read_csv("./data/combustivel/ca-2020-01.csv", sep=";", encoding="utf8")
df_2020_02 = pd.read_csv("./data/combustivel/ca-2020-02.csv", sep=";", encoding="utf8")

df_dataset = pd.concat([df_2020_01, df_2020_02])
df_dataset
```

---
## 3. Pré-processamento

Minimizar ou eliminar problemas existentes em um conjunto de dados, bem como torná-los mais adequados para sua utilização por um determinado algoritmo de aprendizagem de máquina.

## 3.1 Eliminação de atributos irrelevantes

Vamos manter apenas Data da coleta, valor de venda e estado


```python
# remover as colunas id_planta e cidade_origem
data = df_dataset["Data da Coleta"].to_list()
valor = df_dataset["Valor de Venda"].to_list()
estado = df_dataset["Estado - Sigla"].to_list()
produto = df_dataset["Produto"].to_list()

df_dataset = pd.DataFrame()
df_dataset["Data"] = data
df_dataset["Valor"] = valor
df_dataset["Estado"] = estado
df_dataset["Produto"] = produto


# imprimir o dataframe
display(df_dataset.head(n=20))
```

---
## 3.2 Tratamento de atributos com valores ausentes

Verificar se existem atributos com valores ausentes (NaN) na base de dados.

```python
# índices das linhas que contém valores NaN
idxRowNan = pd.isnull(df_dataset).any(axis=1).to_numpy().nonzero()

# imprime apenas as linhas com valores ausentes
display(df_dataset.iloc[idxRowNan])
```

## 4. Análise dos Dados
A análise das características presentes em um conjunto de dados permite a descoberta de padrões e tendências que podem fornecer informações valiosas para compreender o processo que gerou os dados.

```python
# apresenta as principais estatísticas da base de dados para atributos numéricos
df_detalhes = df_dataset.describe()

display(df_detalhes)
```

---

### Agrupamento dos dados
Os preços são coletados de vários postos em cidades e estados diferentes, vamos agrupar por data e estado, calculando a média nacional para uma data específica.

#### Pegamos os valores únicos de data e estado

```python
data_unique = df_dataset["Data"].unique()
estado_unique = df_dataset["Estado"].unique()
produto_unique = df_dataset["Produto"].unique()
display(estado_unique)
display(produto_unique)
```

## Etanol

#### Vamos Encontrar o estado com etanol mais caro e o estado com etanol mais barato

```python
df_etanol = df_dataset[df_dataset["Produto"] == "ETANOL"]
df_etanol.loc[:, 'Valor'] = df_etanol['Valor'].str.replace(',', '.').astype(float).fillna(0.0)

media_por_estado = df_etanol.groupby("Estado")["Valor"].mean()


estado_mais_caro = media_por_estado.idxmax()  # Estado com maior média
valor_mais_alto = media_por_estado.max()

estado_mais_barato = media_por_estado.idxmin()  # Estado com menor média
valor_mais_baixo = media_por_estado.min()

print("Média de valores por estado:")
print(media_por_estado)

print(f"\nEstado mais caro: {estado_mais_caro} com média de {valor_mais_alto:.2f}")
print(f"Estado mais barato: {estado_mais_barato} com média de {valor_mais_baixo:.2f}")
```

#### Vamos agrupar os valores por data e estado

```python
grouped_values=[]
datas=[]
df_etanol = df_dataset[df_dataset["Produto"] == "ETANOL"]
for data in data_unique:
        df_data = df_etanol[df_etanol["Data"]==data]
        
        total=0
        sum_valor=0
        for valor in df_data["Valor"]:
            total+=1
            sum_valor+=float(valor.replace(",", "."))
        if total > 0:
            grouped_values.append(sum_valor/total)
            datas.append(data)
        
            

        
```

```python
df_grouped = pd.DataFrame({"Valor": grouped_values, "Data": datas})

df_grouped['Data'] = pd.to_datetime(df_grouped['Data'], format='%d/%m/%Y')

# Ordenando o DataFrame pela Data
df_grouped = df_grouped.sort_values(by='Data').reset_index().drop("index",axis=1)


df_grouped
```

### Visualizando a média dos preços

```python
# Criando o gráfico de linha
plt.figure(figsize=(10, 5))
sns.lineplot(x=df_grouped['Data'], y=df_grouped['Valor'], marker='o')

# Formatando o gráfico
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Histórico de Valor por Data')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor visualização
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

#### Descrevendo a média nacional

```python
# apresenta as principais estatísticas da base de dados para atributos numéricos
df_detalhes = df_grouped.describe()

display(df_detalhes)
```
